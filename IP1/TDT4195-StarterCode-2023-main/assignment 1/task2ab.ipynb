## Task 2a

```python
def greyscale(im):
    """ Converts an RGB image to greyscale
    
    Args:
        im ([type]): [np.array of shape [H, W, 3]]
    
    Returns:
        im ([type]): [np.array of shape [H, W]]
    """
    #Define height and width of image
    imHeight = im.shape[0]
    imWidth = im.shape[1]

    def convert_to_greyscale (R,G,B):
        return 0.212*R + 0.7152*G + 0.0722*B

    #iterate through every pixel and apply the convertion to greyscale
    for i in range(imHeight):
        for j in range(imWidth):
            im[i][j] = convert_to_greyscale(im[i][j][0], im[i][j][1], im[i][j][2])
    
    return im
```

```python
im_greyscale = greyscale(im)
save_im(output_dir.joinpath("duck_greyscale.jpeg"), im_greyscale, cmap="gray")
plt.imshow(im_greyscale, cmap="gray")
```

```python
def inverse(im):
    """ Finds the inverse of the greyscale image
    
    Args:
        im ([type]): [np.array of shape [H, W]]
    
    Returns:
        im ([type]): [np.array of shape [H, W]]
    """
    #define height and width
    imHeight = im.shape[0]
    imWidth = im.shape[1]
    
     # YOUR CODE HERE
    def intensity_transformation(p):
        return 1-p

    #iterate through every pixel and apply the transformation
    for i in range(imHeight):
        for j in range(imWidth):
            im[i][j] = intensity_transformation(im[i][j][0])
    
    return im
```

```python
im_inverse = inverse(im_greyscale)
save_im(output_dir.joinpath("duck_inverse.jpeg"), im_inverse, cmap="gray")
plt.imshow(im_greyscale, cmap="gray")
```
